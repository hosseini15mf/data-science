### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# There are generally two ways: explicit and implicit interfaces. 
# Explicit way is to create figures and axes and call the methods on them (like object oriented)
# Implicit way is to use pyplot to create and manage the figures and axes and use pyplot functions for plotting.
# Object oriented style is more recommended because you have more control over the plotting via using object oriented style.
# When the project gets more complicated O.O. style gives us more felexibility and management. 
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# the figure is an object that contains everything you can see on the figure window that the plots are displayed.
# Axes is a plot or graph in a figure. It can have x-axis and y-axis, labels, gridlines and data space.
# Axis manages the scaling of the data. examples: x-axis: horizontal axis and y-axis: vertical axis
# Artist is a base class for everything that you can see on a plot in matplotlib. via artists we can have control over the apearance 
# and behaviour of our plots.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2 

fig, ax = plt.subplots()
ax.plot(x, y, label='y = x^2', color='blue')
ax.set_title('Plot of y = x^2 (Object-Oriented Approach)')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
plt.show()

plt.plot(x, y, label='y = x^2', color='green')
plt.title('Plot of y = x^2 (Pyplot Approach)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5))
ax1.scatter(random_data_x, random_data_y, alpha=0.3, color='blue', label='Random Data')
ax1.plot(x, y, color='red', label='y = x^2')
ax1.set_title('Scatter Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.legend()

ax2.bar(names, values, color='green')
ax2.set_title('Bar Plot of Fruit Data')
ax2.set_xlabel('Fruits')
ax2.set_ylabel('Quantity')
plt.tight_layout()
plt.show()
```
